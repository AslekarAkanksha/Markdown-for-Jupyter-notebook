# Markdown for Jupyter Notebook

<h2><p style="color:red">1) <u>Heading</u></p></h2>

For bold headings use :  

1 #   : for title                                                                                                         
2 ##  : for major headings                                                                                                 
3 ### : for subheadings                                                               

<br>

<h2><p style="color:red">2) <u>Text</u></p></h2>

 Markdown for __Jupyter Notebook__

For **bold** text use :

1)  __ text __ or ** text **  (2 underscore without space)...( Here you can't use # for heading, it will bold entire sentence.)
2)  < b > text < /b >  (this is html format)


For **italic** text use :

1)  _ text _ or * text *  (1 underscore)

<br>

<h2><p style="color:red">3) <u>Line breaks</u></p></h2>

 Markdown for <br> Jupyter Notebook

1 : Markdown doesn't make line breaks when you want for better text format, to do this use __< br >__ where you want to break the line.<br> 2 : Also when you want blank space after 1st paragraph ends, then use __< br >__ and hit enter.

<br>

<h2><p style="color:red">4) <u>Indenting</u></p></h2>

> Markdown for Jupyter Notebook 

1 : Use __>__ sign followed by space in the beginning, it will indented any text.

<br>

<h2><p style="color:red">5) <u>Bullets</u></p></h2>

- 

A hyphen (-) followed by one spce 

-  

A hyphen (-) or (*) followed by two space

- *  

A hyphen (-) + space + (*) two space

<br>

<h2><p style="color:red">6) <u>Sub bullet using</u> : <u>tab</u></p></h2>

    - Main point
        - Sub point

Press __tab__ from keyboard then hyphen (-) followed by a space.  Again press enter + tab + (-) + space.

<br>

<h2><p style="color:red">7) <u>Number list</u></p></h2>

1. Jupyter
2. Notebook

Enter a number followed by space

    1. Jupyter
        2. Notebook

Press __tab__ then number followed by space, after text hit enter + tab then number followed by space.

<br>

<h2><p style="color:red">8) <u>Colored Note Boxes</u></p></h2>

<p style="color:orange"><b>Restriction</b> :-Not all Markdown code displays correctly within <i><u><b>div</b></u></i> tags, so review your coloured boxes carefully. For example, to make a word bold, surround it with the <i><u><b>HTML code for bold text instead of the Markdown code.</b></u></i></p>

<br>

<p style="color:blue">- alert-info shows blue color</p>

<div class="alert alert-info">
    <b>Tip :</b> Use blue boxes for tips and notes.
    If it's a note, you don't have to include the word "Note".
    </div>

<br>

<p style="color:yellow">- alert-warning shows brown color letters

<div class="alert alert-warning">
    <b>Example :</b> Use yellow boxes for examlple that are not inside code cells, or use for mathematical formulas if needed.
    </div>

<br>

<p style="color:green">- alert-success shows green color

<div class="alert alert-success">
    <b>Up to you :</b> Use green boxes sparingly, and only for some specific purpose that the other boxes can't cover. For
    example, if you have a lot of related content to link to, maybe you decide to use green boxes for related links from each
    section of a notebook.
    </div>

<br>

<p style="color:red">- alert-danger shows red color

<div class="alert alert-danger">
    <b>Just don't :</b> In general, avoid the red boxes. These should only be used for actions that might cause data loss or
    another major issue.
    </div>

<br>
